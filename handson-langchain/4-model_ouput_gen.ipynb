```python
import os
import time
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
```

```python
load_dotenv()
```

```python
# Initialize models
gemini = ChatGoogleGenerativeAI(model="gemini-1.5-flash-latest", google_api_key=os.getenv("GOOGLE_API_KEY"))
groq = ChatGroq(model="llama3.1-405b-reasoning", api_key=os.getenv("GROQ_API_KEY"))
```

```python
def query_llm(model, prompt):
    start_time = time.time()
    response = model.invoke([HumanMessage(content=prompt)])
    end_time = time.time()
    exec_time = end_time - start_time
    return {"response": response.content, "time": exec_time}
```

```python

prompt = input("Enter your prompt: ")

# Add this check immediately after input
if not prompt.strip():
    prompt = "Default prompt because input was empty."
    print(f"Input was empty. Using default: '{prompt}'")
```

```python
# Step 1: Query Gemini
print("Step 1: Querying Gemini...")
gemini_result = query_llm(gemini, prompt)
print(f"Gemini response: {gemini_result['response']}")
print(f"Gemini time: {gemini_result['time']:.2f}s\n")
```

```python
# Step 2: Query Groq
print("Step 2: Querying Groq...")
groq_result = query_llm(groq, prompt)
print(f"Groq response: {groq_result['response']}")
print(f"Groq time: {groq_result['time']:.2f}s\n")
```
