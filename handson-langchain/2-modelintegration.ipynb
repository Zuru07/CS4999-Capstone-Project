### Streaming and Batch

Basically instead of waiting for the entire output to be generated and then finally the response, the output as it is being processed is printed to the user instead of waiting till the end and then finally giving the output. It is basically every other response you get on aany LLM

```python
for chunk in model.stream("Write me a 200 word paragraph on Artificial Intelligence"):
    print(chunk.text, end="|", flush=True)
```
