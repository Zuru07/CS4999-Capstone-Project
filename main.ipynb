```python
import os
import time
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_groq import ChatGroq

# Load environment variables from .env
load_dotenv() 

# Verify keys are loaded
if not os.getenv("GOOGLE_API_KEY"):
    print("Google API Key missing!")
elif not os.getenv("GROQ_API_KEY"):
    print("Groq API Key missing!")
else:
    print("API Keys loaded.")
```

```python
# Google Gemini (using Gemini 2.0 Flash)
gemini_llm = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash-lite-preview-02-05",
    temperature=0
)

# Groq (using Llama-3.1-8b-instant)
groq_llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0
)

print("Models initialized successfully.")
```

```python
def run_and_time_model(llm, name, prompt_text):
    """
    Runs the LLM with the prompt and calculates execution time.
    """
    print(f" Running {name}...")
    start_time = time.time()
    
    try:
        response = llm.invoke(prompt_text)
        content = response.content
    except Exception as e:
        content = f"Error: {e}"
    
    end_time = time.time()
    duration = end_time - start_time
    
    return content, duration
```

```python
# --- CHANGE THIS PROMPT TO TEST ---
user_prompt = "Explain the concept of recursion to a 5 year old."
# ----------------------------------

print(f" Prompt: {user_prompt}\n")

# Step 1: Run Google Gemini
gemini_output, gemini_time = run_and_time_model(gemini_llm, "Google Gemini", user_prompt)
print(f"    Gemini finished in {gemini_time:.4f}s")

# Step 2: Run Groq
groq_output, groq_time = run_and_time_model(groq_llm, "Groq (Llama 3)", user_prompt)
print(f"   Groq finished in {groq_time:.4f}s")

# Step 3: Compare and Pick Winner
print("\n" + "="*30)
if groq_time < gemini_time:
    winner_name = "Groq"
    winner_output = groq_output
    winner_time = groq_time
else:
    winner_name = "Google Gemini"
    winner_output = gemini_output
    winner_time = gemini_time

print(f"WINNER: {winner_name}")
print(f"TIME:   {winner_time:.4f}s")
print("="*30 + "\n")

print("RESPONSE:")
print(winner_output)
```
